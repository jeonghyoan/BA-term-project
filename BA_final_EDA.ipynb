```python
#Month 의미있는 feature로 변경
d2.rename(columns={'Month':'계절'}, inplace = True)
d2
```

```python
d2['계절']=d2['계절'].astype(int)
index = 0
for m in d2['계절']:
    if(3 <= m <= 5):
        d2['계절'][index] = '봄'
    elif(6 <= m <= 8):
        d2['계절'][index] = '여름'
    elif(9 <= m <= 11):
        d2['계절'][index] = '가을'
    else:
        d2['계절'][index] = '겨울'
    index += 1

d2
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

X_train = d2[['평균기온(℃)','최저기온(℃)','최고기온(℃)','강수량(mm)','배추:면적 (ha)','생산량 (톤)','소비자물가지수','수출액($1000)','수출량(kg)','수입액($1000)','수입량(kg)']]
 
def feature_engineering_XbyVIF(X_train):
    vif = pd.DataFrame()
    vif['VIF_Factor'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
    vif['Feature'] = X_train.columns
    return vif

vif = feature_engineering_XbyVIF(X_train)
print(vif)
```

```python
corr_M = X_train.corr() # 독립변수 상관관계
mask = np.array(corr_M)
mask[np.tril_indices_from(mask)] = False

plt.figure(figsize= (10,8))

sns.heatmap(corr_M, 
            cmap = sns.light_palette("red", as_cmap=True),
            annot = True, 
            mask = mask)

plt.show()
```
